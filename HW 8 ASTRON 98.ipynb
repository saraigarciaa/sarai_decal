```python
#1: 5
def make_sine_wave(x, A, w):
    return A * np.sin(w * x)
x = np.linspace(0, 2 * np.pi, 1000)
amplitudes = np.array([0.5, 1, 1.5, 2, 2.5])
frequencies = np.array([1, 2, 3, 4, 5])
plt.figure(figsize=(10,6))
for A, w in zip(amplitudes,frequencies):
    y = make_sine_wave(x, A, w)
    plt.plot(x, y, label=f'A={A}, w={w}')
    plt.title("Sine Waves with Different Ampltiudes and Frequencies")
    plt.xlabel('x (radians)')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.show
```

```python
penguins = sns.load_dataset("penguins")

fig, axes = plt.subplots(1, 2, figsize=(14,6))
sns.scatterplot(
    data = penguins,
    x = "flipper_length_mm",
    y = "body_mass_g",
    hue = "species",
    ax = axes[0]
)
axes[0].set_title("Flipper Length vs Body Mass")
axes[0].set_xlabel("Flipper Length (mm)")
axes[0].set_ylabel("Body Mass (g)")
axes[0].legend(title="Species")

sns.histplot(
    data = penguins,
    x = "body_mass_g",
    bins = 10,
    ax=axes[1],
)
axes[1].set_title("Distribution of Body Mass")
axes[1].set_xlabel("Body Mass (g)")
axes[1].set_ylabel("Frequency")
```
